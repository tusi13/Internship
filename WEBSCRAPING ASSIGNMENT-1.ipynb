```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
soup=BeautifulSoup(page.content,"html.parser")
soup
```

```python
header_tags = soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6"])

header_texts = [tag.get_text() for tag in header_tags]

df = pd.DataFrame(header_texts, columns=["Header"])

print(df)
```

```python
response=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi')
response
```

```python
team_data = []
table = soup.find('table', class_='table')
rows = table.find_all('tr')

for row in rows[1:11]:
    cells = row.find_all('td')
    team = cells[1].text.strip()
    matches = cells[2].text.strip()
    points = cells[3].text.strip()
    rating = cells[4].text.strip()
    team_data.append([team, matches, points, rating])

df = pd.DataFrame(team_data, columns=['Team', 'Matches', 'Points', 'Rating'])
print(df)
```

Q7) Write a python program to scrape mentioned details from dineout.co.in and make data frame-
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
Name=[]

for i in soup.find_all('div',class_="restnt-info cursor"):
    Name.append(i.text)
    
Name
```

```python
Multiple_location=[] # empty list to store the data

for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    Multiple_location.append(i.text)

Multiple_location
```

```python
price=[] # empty list to store the data

for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text)
    
price
```

```python
image=[]

for i in soup.find_all('img' , class_="no-img"):
    image.append(i['data-src'])
    
image
```

```python
rating=[]

for i in soup.find_all('span',class_="rating-value"):
    rating.append(i.num)
    
rating
```

```python
df=pd.DataFrame({'Titles':titles,'Location':Multiple_location,'Price':price,'Image_url':image})

df
```
